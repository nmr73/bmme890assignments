```python
#whole data --> 891 passengers.
data
```

```python
#histogram of age
data.Age.plot(kind='hist')
plt.xlabel('Age')
plt.title('Histogram of Age for Members on Titanic')
```

```python
#plot graph with Pclass on x axis and Fare on y axis.
plt.plot(data.Pclass,data.Fare,'k+')
plt.xlabel('Passenger Class')
plt.ylabel('Fare Cost ($)')
plt.title('Fare Cost for Each Type of Passenger Class')
```

```python
 #crosstab Survive and Pclass
pd.crosstab(data.Survived,data.Pclass)
```
